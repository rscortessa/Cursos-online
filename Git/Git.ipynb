 ### 5.1 Unstaged files <a id='unstaged'></a>

* ***git checkout*** This command returns to the last commitment that you did.<br>
It can be used when the file is not staged and you made some changes, if the file is staged an you made any change, it returns the original file version that was staged.
 * ***NOTE*** The dash ***-p*** allows to decide return change by change made in the file
* The life is not perfect and sometimes we stage changes that we don't want, for this reason there is a command that allows unstage changes: ***git reset***. We must use it with the alias HEAD and the name of the file that we want to unstage
 * ***ANOTHER NOTE*** You can use the command ***git add**** to add all the changes that were made and new files located in the working tree to the staged area.
 
 
  ### 5.2 Amending commits (commited files) <a id='amending'></a>

* Programmers are humans and made mistakes, if we had troubles with the las commit, we can overwrite it with the dash ***--amending*** in the command ***git commit***.
 * To execute this command first we need to stage our changes, after it will open a window where we can modify the message of the last commit.
 
### 5.3 Rollbacks <a id='rollbacks'></a>

* If you made a commit and after this you see that something goes wrong, it is time to "Rollback", Rollback is a kind of commit that makes the opposite: invert the changes made in a commit. You need to write the command: ***git revert HEAD*** and this will open a window to write the message concerning about this rollback.

### 5.4 Identifying a commit <a id='identifying'></a>
* We have made rollbacks to the last commit that was sent, forthis purpose it was used the "HEAD" alias to specify that the commit to revert was the last but what happened if the error didn't happen in the last commit? for revert any other commit we need to use the ***commit id*** instead of the alias ***HEAD***.
* ***NOTE*** The ***commit id*** is a serial letters that allow to identify a commit done.
## 6. The Branches <a id='branches'></a>

### What is a branch? <a id='whatis'></a>

* It is defined as *"A pointer to a particular commit"* (Represent an independent line of development).<br> 


* The default branch that git creates for you when a repository is initialized is called ***master***


* Having different branches allows you experiment with new ideas or correct some mistakes in the project without affect all the work saved in the branch HEAD. <br> Think about this as different notebooks where you develop ideas of a project, these books are branches and the notebook where you write your final version of your work is the HEAD.
### Creating new branches <a id='newbranches'></a>

* We can use the ***git branch*** command to list, create, delete and manipulate branches. <br> Running ***git branch*** only shows all the branches created in the repository. <br> The command ***git branch new branch*** creates a new branch in the repository and with the command  ***git checkout new branch***  we can also switch to this branch, this steps we can make them in a single command ***git checkout -b new branch***.
* Switching to another branch also changes the working tree and the commit history according to the last commit of the branch, if you wish delete a branch the command ***git checkout -d name-of-branch*** allows you to do that but if there are changes in the branch to delete that aren't in the HEAD, git will show a screen with a warning
### Merging <a id='Merging'></a>

* ***Merging*** is the term that Git uses for combining branched data and history together<br> We use the command ***git merge-name-of-branch*** to combine the changes made in this branch with the HEAD, we will have problems if the changes made in these two branches diverge creating a ***merge conflict***
* solving merge conflicts: if we have a merge conflict it is necessary to be solved, in that case we need to find where is located the conflict, for that reason is very useful to use ***git status*** to check the state of the repository and to find out hints, once found the conflict, we can edit the file to solve the problem after we need to add the files and commit them.  
* if we can abort the merge we use *** git merge --abort***
* the command ***git log --graph --oneline*** shows a graph about the commit history in the branches, it is very useful to understand merge conflicts.

# Working with remotes <a id='Remotes'></a>

## What is GitHub?<a id='GitHub'></a> 

GitHub can be described as a remote repository hosting service for Git.

## Basic interaction with GitHub<a id='basicinteraction'></a> 


* First you need a GitHub account, after you create the repository and give it a name and a description.
* After that we create a copy of the repository in the computer, for this you have to use the command ***git clone*** and add the link of your repository.
* We can interact with the local repository on your machine but if you change something, you need to send this information to the repository localized in GitHub, to do that the command ***git push***
* If you want to update your repository on your machine you must type the command ***git pull***

## What is a remote?<a id='remote'></a> 

* With GitHub we can interact with remote repositories with our local repository, it allows work with other people and the same time, these people work on theirs local repositories and updates theirs changes in the remote repository, you need to care about what changes were made in the remote repository pull them and merge them with your local repository after that you can send it to the remote with push.
* One thing you must know is that in the remote repository can be created branches as well as the local repository allows, to access to one remote branch you need to pull it, merge it with the local branch then push back to its origin.
* In the remote repository there is a master branch called origin which is update according to the master branch of the local repository.
* These commands can be useful: ***git remote -v***,***git branch -r*** and the most important ***git remote show origin***, the last command give us the information about the local and remote repo.
## Fetching new changes <a id='fetching'></a> 

* We have seen how pull updates to the local repository and push to the remote repository, now there is a command that allows do the same as pull, but it doesn't merge with our local repository, this command is ***git fetch***, with ***git fetch*** we can download all the changes made in the remote without merge in the local repo, the master in the remote repository is named as ***origin/master***, so knowing its name you can merge it with the local repository.
* Also you can use ***git fetch*** to review the changes made in the remote repository

## Updating the local repository <a id='updating'></a> 

* It is known the command ***git pull*** allows fetch and merge the contents of the local and remote repo, if there are more than one remote repo we can access to them only typing ***git checkout name-of-branch*** and we will working locally in the remote repo. 
* if we want the contents of the remote branches without merging we can use ***git remote update***
